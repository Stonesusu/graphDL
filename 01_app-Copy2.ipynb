```python
import json
def json_resolver(x):
    columns = ['customer_id','android_id','loan_app_id','trans_time','create_time','appName','isSystemInstalled',
               'firstInstallTime','lastUpdateTime','packageName','versionCode','versionName']
    results = []
    tmp = json.loads(x['applist'])
    for ele in tmp:
        appName = ele.get('appName',np.nan)
        isSystemInstalled = ele.get('isSystemInstalled',np.nan)
        firstInstallTime = ele.get('firstInstallTime',np.nan)
        lastUpdateTime = ele.get('lastUpdateTime',np.nan)
        packageName = ele.get('packageName',np.nan)
        versionCode = ele.get('versionCode',np.nan)
        versionName = ele.get('versionName',np.nan)
        result = [x['customer_id'],x['android_id'],x['loan_app_id'],x['trans_time'],x['create_time']]
        result.append(appName)
        result.append(isSystemInstalled)
        result.append(firstInstallTime)
        result.append(lastUpdateTime)
        result.append(packageName)
        result.append(versionCode)
        result.append(versionName)
        results.append(result)
    return pd.DataFrame(results,columns=columns)
```

```python
import time
import gc
import multiprocessing
start = time.time()
pool = multiprocessing.Pool(processes=10)
executeResults={}
for idx in range(tmp.shape[0]):
    key = str(idx)
    executeResults[key] = pool.apply_async(func=json_resolver,args=(tmp.iloc[idx],))
pool.close()
pool.join()
results = pd.DataFrame()
for k,value in executeResults.items():
    result = value.get()
    results = pd.concat([results,result])
del executeResults
gc.collect()

end = time.time()
print("Execution Time: ", end - start)
```

```python
import re
#appname 特殊处理
special_character="[-|&|@|(|)|+|!|_|:|\n|']"
def appnameClean(appname):
    # 剔除特殊字符
    appname=re.sub(special_character,"",str(appname))
    # 去除空格
    appname=appname.replace(' ','')
    #转小写
    appname=appname.lower()
    return appname
```

```python
def app_jaccard_similarity(m,n,a,b):
    intersection = len(np.intersect1d(a,b))
    union = len(np.union1d(a,b))
    jaccard = 1.0*intersection/union if union>0 else -9999976.0
    return [m,n,intersection,union,jaccard]
```

```python
import time
tmp = df1

start = time.time()

pool = multiprocessing.Pool(processes=10)
executeResults = {}
for i in range(tmp.shape[0]):
    for j in range(tmp.shape[0]):
        if j<=i:
            continue
        m = str(tmp.iloc[i]['customer_id'])
        n = str(tmp.iloc[j]['customer_id'])
        a = tmp.iloc[i]['appNamelist']
        b = tmp.iloc[j]['appNamelist']
        executeResults[m+'_'+n] = pool.apply_async(func=app_jaccard_similarity,args=(m,n,a,b))
pool.close()
pool.join()
results = []
for k,value in executeResults.items():
    tmp = value.get()
    results.append(tmp)

end = time.time()
print("Execution Time: ", end - start)
```
